```python
import pandas as pd
data=pd.read_csv('/kaggle/input/phishing/custom_phishing_dataset.csv')
print(data.head())
print("rows and coloumns:",data.shape)
print("missing values:\n",data.isnull().sum())
print("target distribution:\n",data['Is_Phishing'].value_counts())
```

```python
data.dropna(inplace=True)
X=data[['Have_IP','Have_At_Symbol','URL_Length','Is_Https']]
y=data['Is_Phishing']
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score



#Split data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

#Train model

model = LogisticRegression()

model.fit(X_train, y_train)

#Evaluate

y_pred=model.predict(X_test)

print("Accuracy on test set:", accuracy_score(y_test, y_pred))
```
